#### 1.  Задание

* Даны два вектора в трехмерном пространстве: (20,20,20) и (0,0,-20)
* Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. (в программе)

```python
def vlen(v):
    sm = 0
    for x in v:
        sm += x**2
    return math.sqrt(sm)

vlen([20,20,20])

# другой способ
# a = np.array([20, 20, 20])
# math.sqrt(np.sum(a**2))
```

#### 3.  Задание (в программе)

* Напишите код на Python, реализующий построение графиков:    

1. окружности

```python
R = 2
x = np.linspace(-R, R,1000)
fig, ax = plt.subplots(1,1)
ax.plot(x, np.sqrt(R**2 - x**2), color="blue")
ax.plot(x, -np.sqrt(R**2 - x**2), color="blue")
fig.set_size_inches(3, 3)
plt.show()
```

2. эллипса

```python
a=2
b=5
x = np.linspace(-2, 2,1000)
y = np.sqrt(b**2 - x**2 * b**2 / a**2)

fig, ax = plt.subplots(1,1)
ax.plot(x, y, color="blue")
ax.plot(x, -y, color="blue")
plt.show()
```

3. гиперболы. 

```python
a=1
b=2
x = np.linspace(1, 2,1000)
y = np.sqrt(x**2 * b**2 / a**2 - b**2)

fig, ax = plt.subplots(1,1)
ax.plot(x, y, color="blue")
ax.plot(x, -y, color="blue")
plt.show()
```

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
```

* 1) Нарисуйте трехмерный график двух параллельных плоскостей.

```python
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(X, Y)
Z1 = 2*X + 5*Y
Z2 = Z1 + 25
ax.plot_wireframe(X, Y, Z1)
ax.plot_wireframe(X, Y, Z2, color="red")
show()
```

* 2) Нарисуйте трехмерный график двух любых поверхностей второго порядка.

```python
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(X, Y)
Z = X**2 + Y**2
ax.plot_surface(X, Y, Z)
show()
```
